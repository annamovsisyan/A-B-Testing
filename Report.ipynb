```python
from bandit import Bandit, EpsilonGreedy, ThompsonSampling, Visualization, comparison
```

```python
Bandit_Reward = [1, 2, 3, 4]
NumberOfTrials = 20000
```

```python
(cumulative_avg_reward_e, cumulative_total_reward_e, 
    cumulative_regrets_e, greedy_bandits,
    bandit_choices_e, rewards_e, suboptimal_pull_count) = EpsilonGreedy(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

```python
Visualization().plot1(NumberOfTrials, cumulative_avg_reward_e, greedy_bandits, 'Epsilon Greedy')
```

```python
(cumulative_avg_reward_t, cumulative_total_reward_t, 
     cumulative_regrets_t, thompson_bandits, 
     bandit_choices_t, rewards_t) = ThompsonSampling(Bandit).experiment(Bandit_Reward, NumberOfTrials)
```

```python
Visualization().plot1(NumberOfTrials, cumulative_avg_reward_t, thompson_bandits, 'Thompson Sampling')
```

```python
Visualization().plot2(cumulative_total_reward_t, cumulative_total_reward_e, cumulative_regrets_t, cumulative_regrets_e)
```

```python
comparison(NumberOfTrials, cumulative_avg_reward_e, cumulative_avg_reward_t, 
           cumulative_total_reward_e, cumulative_total_reward_t, 
           cumulative_regrets_e, cumulative_regrets_t, 
           thompson_bandits)
```

```python
EpsilonGreedy(Bandit).report(NumberOfTrials, greedy_bandits, 
                             bandit_choices_e, rewards_e, 
                             cumulative_regrets_e, suboptimal_pull_count, 
                             'EpsilonGreedy')
```

```python
ThompsonSampling(Bandit).report(NumberOfTrials, thompson_bandits, 
                             bandit_choices_t, rewards_t, 
                             cumulative_regrets_t, 
                             algorithm='ThompsonSampling')
```
